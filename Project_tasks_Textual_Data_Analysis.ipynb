# Task 1



*   SQuAD: 100,000+ Questions for Machine Comprehension of Text

This paper illustrates the difficulties that machines encounter when it comes to reading comprehension, including the need to comprehend natural language and possess knowledge of the outside world. Based on interannotator agreement, the paper compares human performance with that of the best model on the SQuAD dataset.
The outcome of this paper is what I found most fascinating. Exact match and F1 score were the two criteria used to compare model and human performance on the SQuAD dataset. Results indicate that on both the metrics, humans outperformed the models. It is noted that most of the mismatch between human and machine performance occurred due to the inclusion or exclusion of non-essential phrases. This suggests that humans may have a better understanding of the language context and complexity, allowing them to identify the most appropriate answer even if it does not match the exact phrasing of the question. Additionally, humans may be able to use common sense and external knowledge to answer questions, while the machine relies solely on the provided text.
Interannotator agreement was used to evaluate human performance, which is another intriguing aspect. So, humans receiving a high score suggests a high level of interannotator agreement. As a result, the responses of human annotators who independently annotate or evaluate the same set of data revealed a significant amount of resemblance.
Finally, another interesting part of this research is the data collection process. To develop questions and responses on the information in each paragraph, crowdworkers were employed. Employees were obliged to spend 4 minutes on each paragraph and were paid $9 per hour for the total number of hours it took to finish the piece. Tasks included a sample paragraph, as well as good and terrible questions and answers on that paragraph, to serve as a guide for the workers. The staff members were urged to formulate their own queries rather than paraphrasing sentences from the paragraph. This is a key factor in how this dataset has such a large amount of material.

* Know What You Don’t Know Unanswerable Questions for SQuAD

This paper is about the SQuAD 2.0 dataset. Unlike the original SQuAD dataset, it contains certain unanswerable questions in order to make the dataset more difficult for the models. An intriguing aspect of this paper is the authors' emphasized on two criteria for the proposed SQuAD 2.0 dataset, relevance and existence of probable replies, which are specific to unanswerable questions. These criteria ensure that the unanswerable questions are challenging and require models to have a deeper understanding of the context and the language.
Another interesting fact similar to the SQuAD dataset is, here also crowdworkers were employed in the development of SQuAD 2.0 dataset, however this time, there were actually two groups of crowdworkers. The workers were instructed to ask up to five questions that could not be answered by reading the paragraph alone, while citing specific paragraph elements and making sure that a plausible response was offered. The authors paid additional crowdworkers to respond to every query in the SQuAD 2.0 development and test sets in order to verify that the dataset is error-free.
Finally, the most fascinating aspect of this study is using plausible answers as distractors. Here, the measurement was based on how frequently systems were tricked into responding to the implausible queries with the persuade replies offered by crowdworkers. Each of the models as well as the human answerers made approximately half of their false positive errors due to plausible answers. The dataset's plausible replies indeed served as distracting possibilities that could be confused for the right response.

```python
dset = dset.sort("timestamp")
print(dset)
```

```python
pipline = transformers.pipeline(task="feature-extraction",model="bert-base-cased",return_tensors=True,device=0)

en_embedded = pipline(transformers.pipelines.pt_utils.KeyDataset(dset['train'], "title"), batch_size=64) #, truncation="only_first")
# we tried to remove the truncation = 'only_first' argument because our concern is not to retain the most informative part of the input sequence, but the whole article

# encode the question as input to the model
question = 'How long was the prison sentence in the 4th Oulu child sex abuse case'
#question = 'In what cities did rents rise across Finland'
question_embedding = transformers.pipeline("feature-extraction", model="bert-base-cased", return_tensors=True, device=0)(question)
```

```python
import torch 

en_embedded_pooled=[torch.mean(elem,axis=1) for elem in en_embedded]
all_en=torch.vstack(en_embedded_pooled)
```

```python
question_embedded_pooled=[torch.mean(question_embedding,axis=1)]
all_question=torch.vstack(question_embedded_pooled)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

sims = cosine_similarity(all_question.numpy(), all_en.numpy())

# set the number of top articles to retrieve
K = 5

# get the indices of the top K articles with highest similarity
top_k_indices = np.argsort(sims[0])[::-1][:K]

# retrieve the top K articles
top_k_articles = [dset['train']['text'][i] for i in top_k_indices]
```

To retrieve the top K articles most similar to the question, we used the argsort method of numpy to get the indices of these articles based on the cosine similarity scores. We were then able to retrieve the article text for these top K articles and use a QA model to find the answer.

To handle long articles, we split the articles into smaller sections of a fixed length (512 as it is the maximum length of input that can be accepted by the model) and feed these sections to the QA model one at a time. We then loop through each article and its sections to find the section that gives the highest score from the QA model, and finally we output the answer with the highest score.

```python
# get the answer to the question from the top K articles

best_answer = None
best_score = 0.0

# split articles into smaller sections
section_length = 512
#sections = []

for article in top_k_articles:
    n_sections = len(article) // section_length + 1
    for i in range(n_sections):
        section_start = i * section_length
        section_end = (i + 1) * section_length
        section = article[section_start:section_end]
        answer = qa_pipline(question=question, context=section)
        # choose the answer with the highest score
        if answer["score"] > best_score:
            best_answer = answer["answer"]
            best_score = answer["score"]

print('Question:', question)
print("Answer:", best_answer)
print(f"Confidence level: {best_score:.2%}") 
```

The results we got from this task are a bit confusing, since looking at the score (which looks promising) would let us believe that the answer is correct with high confidence. However, the issue is that the answer is almost never what we actually expect. As a conclusion, we couldn't tell if the issue is with the model itself, or is it because of our choice of the question.
